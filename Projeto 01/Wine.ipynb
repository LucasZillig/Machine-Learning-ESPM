# Carregamento do dataset 

```python
from sklearn import datasets
import pandas as pd
import numpy as np

wine = datasets.load_wine()
```

# Pré Processamento dos dados

<h2>Foi utilizado o a função "np.c" para concatenar e assim finalizar a conversão de sklearn.utils.Bunch para Pandas DataFrame</h2>

```python
df = pd.DataFrame(data = np.c_[wine.data, wine.target], columns = wine['feature_names'] + ['target'])
df.head()
```

<h3>Foi utilizada a função replace presente no Pandas Dataframe para substituir os valores numéricos, para valores categóricos na coluna target: class_0(0), class_1(1), class_2(2)</h3>

```python
df.target.replace(df['target'].unique(), wine['target_names'], inplace=True)
```

```python
df["color_intensity"].iloc[10:18] = np.nan
```
